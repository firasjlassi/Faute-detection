```python
import tensorflow as tf
from tensorflow import keras
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.utils import to_categorical
from numpy import argmax
```

```python
x=np.zeros((85,96,96,3),dtype=np.float32)
ye=np.zeros((85,1),dtype=np.float32)
```

```python
p=0
for filename in os.listdir('/content/drive/MyDrive/tackle-football'):
    for takle in os.listdir('/content/drive/MyDrive/tackle-football'+'/'+filename):
        img=cv2.imread('/content/drive/MyDrive/tackle-football'+'/'+filename+'/'+takle)
        x[p]=cv2.resize(img,(96,96))
        ye[p]=int(filename)
        p=p+1
y = to_categorical(ye)
print(y)
```

```python
x=x/255.0
```

```python
print(x)
```

```python
#taw bech n9asmou train w test bl skelarn ashel meli n9asmouhom b numpy kif 3malna bkry
from sklearn.model_selection import train_test_split
x_train , x_test, y_train  , y_test =train_test_split(x, y , test_size=0.2 , random_state=42)
```

```python
x[0]
```

```python
model=tf.keras.Sequential([
                           

tf.keras.layers.Input((96,96,3)),
tf.keras.layers.Conv2D(32,kernel_size=(3,3),activation='relu'),
tf.keras.layers.MaxPool2D((2,2)),


tf.keras.layers.Conv2D(16,(3,3),activation='relu'),
tf.keras.layers.MaxPool2D((2,2)),

tf.keras.layers.Conv2D(8,(3,3),activation='relu'),
tf.keras.layers.MaxPool2D((2,2)),

tf.keras.layers.Conv2D(2,(3,3),activation='relu'),


tf.keras.layers.Flatten(),
tf.keras.layers.Dense(256,activation='relu'),
tf.keras.layers.Dropout(0.5),
tf.keras.layers.Dense(2, activation='softmax') ])
```

```python
model.compile(loss='categorical_crossentropy', 
              optimizer='adam', 
              metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,6,epochs=500)
```
